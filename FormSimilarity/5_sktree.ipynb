Wiki Matrix to Tree

```python
import sys
if sys.version_info[0] < 3: 
    from StringIO import StringIO
else:
    from io import StringIO

WikiData = StringIO(
"""a,b,c,d,e,f,g
a,0,5,9,9,8,0,0
b,5,0,10,10,9,5,5
c,9,10,0,8,7,9,9
d,9,10,8,0,3,9,9
e,8,9,7,3,0,8,8
f,0,5,9,9,8,0,0
g,0,5,9,9,8,0,0
    """)

wdf = pd.read_csv(WikiData, sep=",")

wdf
```

```python
dm = DistanceMatrix(wdf, ['a','b','c','d','e','f','g'])



tree = nj(dm)
str_tree = tree.ascii_art()

text_file = open("result/tree_wiki.txt", "w")
text_file.write(str_tree)
text_file.close()

print(str_tree)
```
